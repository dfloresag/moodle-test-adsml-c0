# Submit your first project

Now please use the previously learned skills to actually populate a repository that we created for you, and that you will get invited to once you click on the `Next` button on the platform. You will then be following the very same steps you have done before, except that you will not have to create a new repository yourself.

In order for you to submit the project please do the following:  

1. Create a file named `hello.txt` with some text inside in the local repository and commit it,  
2. Make a **change to the same file** and commit again,  
3. Push your changes to GitHub,  
4. Click on “submit the project” on our platform.  

What should my text file contain? Anything is fine. For example, in the **first commit** you could introduce yourself and tell us a bit about your background in data analytics, data science and coding. In the **second commit** you could add information about what you expect to gain from our program and how you hope to apply your new learned skills once you have completed this program. This is entirely voluntary, but it helps us support you better over the 4 courses ahead of you. Of course, you can commit any other text, make changes, re-commit and push everything to your repository.

Once you have pushed your changes, you can click on “submit the project” on our platform.

On the project page, you have also the possibility of reaching out to us in a direct way, through “Comments”. This is a communication channel that is private between you and us, and you can use it in addition to questions in the units, for more personal communication with the instructors.

::: {.callout-note}

If for any reason, you cannot use Git or Github (e.g. you cannot install it) Github offers the direct upload
option on the repository webpage. In this case, you will basically not use Git and just use Github as a file
exchange platform, just like you would use Dropbox. This is to be considered as a last resort, as learning a
bit of Git and Github workflows is actually important to work on data science projects.

:::
