# Installing Git and getting a Github account

## 1 - Downloading Git

Before starting, you may need to download **git** from this [web
page](https://git-scm.com/downloads). Depending on the computer you are
using, you might already have it installed or not.

### macOS

On most of the recent Mac computers, Git will already be installed. You
don’t have to do anything to install it.

To check if it is installed, you could go to your terminal, and then you
can check that Git is installed by typing the following (after the `$`
sign):

```text
computername:~ username$ git --version
```

You should get an output similar to this:

```text
git version 2.33.1 
```

Note that the version number and exact spelling can change depending on
your release of macOS.

If you don’t get any message printed in your `Terminal`, a blank line, or
even an error message, it just means that Git is not yet installed on
your Mac. You should head to <https://git-scm.com/downloads> to get the
latest version. You can read there about installing Git. Leave all the
options to default. When the setup is finished, **restart your
computer**. When it is done, proceed with the same test as described
above.

### Windows

Windows does not come with Git pre-installed. You will have to download
it and install it yourself. Visit <https://git-scm.com/downloads> and
select the version for your operating system. Leave all the options to
default. When the setup is finished, **restart your computer**.

## 2 - Create an account on Github

Now that you have **git** on your computer, you could use it for your
own work. In order to share it with collaborators or with us, in the
case of your course projects, you will need to have a GitHub account on
<https://github.com/>. Please sign up on the website, using your email
address.

## 3 - Download Github Desktop

To facilitate the workflow, install [GitHub
desktop](https://desktop.github.com) on your machine.

::: {.callout-note}

If you are using a work computer for which you do not have the rights,
ask your IT service for the possibility to install Git and Github. If
you cannot, you will be able to use drag and drop features on Github
itself, but it is **not recommended**.  
:::

## 4 - Configuring Git

Once you have Git and Github desktop installed, and your own Github
account, you will need to proceed in connecting them all together. First
login into Github desktop using your Github credentials, then proceed by
configuring Git, to correctly communicate with Github. In order to do
so, you will need to follow the instructions found on this
[page](https://docs.github.com/en/desktop/installing-and-configuring-github-desktop/configuring-and-customizing-github-desktop/configuring-git-for-github-desktop).
