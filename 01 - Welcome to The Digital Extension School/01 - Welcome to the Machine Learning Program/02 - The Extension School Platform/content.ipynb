# A course overview

The didactic material is divided into four courses:

1. **Introduction to Data Science with Python**

2. **Applied Data Analysis**

3. **Applied Machine Learning 1**

4. **Applied Machine Learning 2**

In **Course 1**, you will be introduced to how to use Python for data science, how to perform basic data manipulation and exploration as well as a little refreshment of some useful mathematical concepts. **Course 2** dives deeper into using data from different sources and coming in different structures. You will also learn more advanced data analysis and data visualization concepts. By the end of the first two courses, you will indeed be able to perform a thorough exploratory data analysis of different messy datasets.  

In **Courses 3 and 4**, you will then deep dive into machine learning. You will build models in Python, from basic linear regression to neural networks, and apply machine learning concepts to real-life examples. **Course 3** will use regression problems to introduce key concepts and approaches of the machine learning pipeline using the sklearn package. These are expanded further in **course 4**, where you will work on classification tasks and unsupervised learning problems before diving into some deep learning with Tensorflow.

::: {.callout-note}

At the end of each course, you are required to complete and submit a
project, that we will thoroughly grade and give you feedback on. The
projects are designed for you to consolidate what you have learned in
each course and put it into practice. 
::: 

Once you finish Course 4, you will then reach the **Capstone** phase. At this point, you will be able to apply what you have learned to any dataset that might be of interest to you or the company you work for. This is your opportunity to show off all your new skills.

In your learning journey, you are actually supported by us, your instructors:
Amir, Christian, Michael, Mohamad, and Panagiota.

The study material is organized in **courses**, **subjects** and **units**. At the bottom of each unit, you can find answers to questions that other learners have already asked.

Each unit is associated with an approximate time which is an indication of how long it would typically take you to finish this unit (see on the top right of your screen). But remember everyone has their own learning approach. The menu on the right shows you in which unit you are in the current course and subject. It also shows you the completion rate of the subject and the course. Please note that courses, as well as subjects and units, vary in length. For example, the first course is a bit shorter than the others as it introduces the general topics, deals with the setup of the working environment, and revises a view concept. In the machine learning courses, the units tend to be longer as we generally need more steps and code to demonstrate more complex concepts or train models.

![](images/illustrations_003.png){width='800'}

Once you move on in the course, you will have access to more units and you will always be able to go back to previous units. For instance, once you finish Course 1 you will be able to navigate through the entire Course 1, without limit.

::: {.callout-note}

The material also remains accessible to you when you successfully finished the program. 
So you will always be able to use it as a reference in the future.
